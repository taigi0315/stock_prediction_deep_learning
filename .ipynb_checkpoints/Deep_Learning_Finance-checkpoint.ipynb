```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from pandas_datareader import data
import quandl
import datetime
```

```python
start_date = '2015-01-01'
end_date = '2017-01-01'
index_list = ['^GSPC', '^DJI', '^IXIC', '^RUT']
moving_windows = [10, 20, 40, 80, 160]
```

```python
def get_index_data(tickers, start_date, end_date):
    data_list = []
    vol_col_data = []
    column_names = []
    
    for t in tickers:
        data_list.append(data.DataReader(t, 'yahoo', start_date, end_date))
        column_names.extend([t+'_vol', t+'_close'])
                             
    for dat in data_list:
        vol_col_data.append(dat[['Volume', 'Adj Close']])
    result = pd.concat(vol_col_data, axis=1)
    result.columns = column_names
    return result

def get_target_data(ticker, start_date, end_date):
    target_data = data.DataReader(ticker, 'yahoo', start_date, end_date)
    target_data = target_data[['Volume', 'Close']]
    return target_data

def build_moving_avg(target_data, windows):
    result = []
    for win in windows:
        moving_avg = target_data.rolling(window=win).mean()
        moving_avg.columns = [str(win)+'_avg_vol', str(win)+'_avg_clo']
        result.append(moving_avg)
    return pd.concat(result, axis=1)

def get_y_value(target_data, date_length):
    y = target_data[date_length:]['Close']
    y.rename('Y')
    return y
```

```python
target_data = get_target_data('AAPL', start_date, end_date)
moving_avg_data = build_moving_avg(target_data, moving_windows)
index_data = get_index_data(index_list, start_date, end_date )

Y = get_y_value(target_data, 5)
```

```python
figure(num=None, figsize=(8,6))
plt.plot(target_data['Close'], label='AAPL')
plt.plot(moving_avg_data['20_avg_clo'], label='20_days_avg')
plt.plot(moving_avg_data['40_avg_clo'], label='40_days_avg')
plt.legend(loc='upper right')
plt.show()
```
